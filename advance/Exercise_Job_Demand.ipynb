```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df.job_country == 'United States'].copy()
df_us['job_posted_month'] = df_us['job_posted_date'].dt.month #but this is gonna show numbers, so we gonna use the code below to show January..., etc.
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
df_us.head()
```

```python
df_us_pivot = df_us.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot
```

```python
df_us_pivot = df_us_pivot.reset_index()
df_us_pivot['job_posted_month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B')
df_us_pivot['job_posted_month_no'] = df_us_pivot['job_posted_month_no'].dt.month
df_us_pivot.set_index('job_posted_month_no',inplace=True)
df_us_pivot.sort_index(inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot
```

```python
df_us_pivot.plot(kind='line')
```

```python
top_3 = df_us['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
top_3_pivot = df_us_pivot[top_3]
top_3_pivot.plot(kind='line')
plt.title('Monthly Job Posting for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.xticks(rotation = 45)
plt.legend
plt.show()
```
