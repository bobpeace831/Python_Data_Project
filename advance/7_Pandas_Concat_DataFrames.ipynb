## Lecture Note

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
months = df['job_posted_month'].unique()
dict_months = {}
for month in months:
    dict_months[month] = df[df['job_posted_month'] == month]
# dict_months = {month: df[df['job_posted_month'] == month] for month in months}
df_q1 = pd.concat([dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]], ignore_index = True)
df_q1['job_posted_month'].value_counts().plot(kind='bar')
```

## Homework

```python
jobs = df['job_title_short'].unique()
data_job = {}
for job in jobs:
    data_job[job] = df[df['job_title_short'] == job]
data_scientist_df = data_job['Data Scientist']
data_analyst_df = data_job['Data Analyst']
combined_df = pd.concat([data_scientist_df,data_analyst_df], ignore_index=True)
combined_df['job_title_short'].value_counts().plot(kind='bar')
```

```python
months = df['job_posted_month'].unique()
df_months = {}
for month in months:
    df_months[month] = df[df['job_posted_month'] == month]
oct_df = df_months['Oct']
nov_df = df_months['Nov']
dec_df = df_months['Dec']
combined_df = pd.concat([oct_df, nov_df, dec_df], ignore_index=True)
combined_df['job_posted_month'].value_counts().plot(kind='bar')
```

```python
full_time_df = df[df['job_schedule_type'] == 'Full-time']
part_time_df = df[df['job_schedule_type'] == 'Part-time']
combined_df = pd.concat([full_time_df, part_time_df], ignore_index=True)
combined_df['job_schedule_type'].value_counts().plot(kind='bar')
combined_df.groupby('job_schedule_type').size()
```
